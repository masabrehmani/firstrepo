```python
fao_pk2 = pd.read_csv("FAOSTAT_data_6-18-2022.csv")
fao_pk2.head()
```

```python
data_of_pk = fao_pk2.drop(["Domain", "Area Code", "Element Code", "Item Code", "Item", "Domain Code", "Unit", "Flag Description" ], axis=1)
data_of_pk.head()
```

```python
sns.violinplot(data=data_of_pk, 
x = "Flag", y="Year", hue="Element")
plt.show()
```

```python
sns.histplot(data = data_of_pk,
x = 'Value', y="Year", hue="Element")
plt.title("Pakistan ka plot")
plt.show()
```
